```python
# Import necessary libraries
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Load the dataset
file_path = "C:\\Users\\Nishuraj Saikia\\OneDrive\\Documents\\recommendation\\FEdataset.csv"
df = pd.read_csv(file_path)

# Convert necessary columns to numeric
df['DB_Value'] = pd.to_numeric(df['DB_Value'], errors='coerce')
df['Tempo'] = pd.to_numeric(df['Tempo'], errors='coerce')
df['Camelot'] = pd.to_numeric(df['Camelot'], errors='coerce')

# Fill missing values only for numeric columns
df.fillna(df.select_dtypes(include=['number']).median(), inplace=True)

# Verify if there are any remaining NaN values
if df.isnull().sum().sum() > 0:
    print("⚠️ Warning: Some NaN values still exist. Filling with 0.")
    df.fillna(0, inplace=True)

# Extract Song Names
song_names = df['Songsname']

# Extract Features (all numerical columns except "Songsname")
features = df.drop(columns=['Songsname'])

# Verify data types
print("\n🔍 Checking data types:\n", features.dtypes)

print("✅ Data Loaded & Prepared Successfully!")

# ------------------------ Step 1: Train the KNN Model ------------------------
knn = NearestNeighbors(n_neighbors=6, metric='euclidean')  # 6 because one result will be the song itself
knn.fit(features)

print("✅ KNN Model Trained!")

# ------------------------ Step 2: Test the Model (Find Similar Songs) ------------------------
def recommend_songs(song_name, df, knn, n_recommendations=5):
    """
    Recommend similar songs based on a given song name using the trained KNN model.
    """
    if song_name not in df['Songsname'].values:
        print("❌ Song not found in dataset!")
        return []
    
    # Find the index of the given song
    song_index = df[df['Songsname'] == song_name].index[0]
    
    # Get the song's features
    song_features = features.iloc[song_index].values.reshape(1, -1)
    
    # Find the nearest neighbors
    distances, indices = knn.kneighbors(song_features, n_neighbors=n_recommendations + 1)
    
    # Get recommended songs (excluding the input song itself)
    recommended_songs = [df.iloc[i]['Songsname'] for i in indices.flatten()[1:]]
    
    return recommended_songs

# ------------------------ Step 3: Test the Recommendation Function ------------------------

# Pick a song to test (Replace 'G1' with any song name from your dataset)
test_song = 'G1'  

recommended_songs = recommend_songs(test_song, df, knn)

# Print Recommended Songs
print(f"\n🎵 Songs similar to '{test_song}':")
for song in recommended_songs:
    print(f"   - {song}")
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

# 🔹 Load dataset (Assuming `df` is already loaded in your script)
# df = pd.read_csv("your_dataset.csv")  # Uncomment if needed

# 🔹 Check for missing values
print("🔍 Checking missing values:\n", df.isnull().sum())

# 🔹 Handle missing values (Replace NaNs with column median)
imputer = SimpleImputer(strategy="median")
df.iloc[:, 1:] = imputer.fit_transform(df.iloc[:, 1:])

# 🔹 Scale the features (excluding 'Songsname')
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.iloc[:, 1:])

# 🔹 Train the KNN model
knn = NearestNeighbors(n_neighbors=6, metric="euclidean")  # 6 (5 recommendations + 1 original)
knn.fit(scaled_features)

# 🔹 Function to recommend songs
def recommend_songs(song_name, df, knn, scaled_features):
    # 🔍 Find the index of the song
    if song_name not in df["Songsname"].values:
        print(f"❌ Song '{song_name}' not found in the dataset.")
        return
    
    song_index = df[df["Songsname"] == song_name].index[0]

    # 🔍 Find similar songs
    distances, indices = knn.kneighbors([scaled_features[song_index]])

    # 🔹 Get recommended songs (excluding itself)
    recommended_songs = df.iloc[indices[0][1:], 0].tolist()
    
    print(f"🎵 Recommended songs for '{song_name}':\n", recommended_songs)

# 🔥 Example Usage
recommend_songs("K136", df, knn, scaled_features)
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Load the dataset
file_path = r"C:\Users\Nishuraj Saikia\OneDrive\Documents\recommendation\FEdataset.csv"
df = pd.read_csv(file_path)

# Convert necessary columns to numeric (in case of any non-numeric values)
numeric_columns = ['DB_Value', 'Tempo', 'Camelot']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Fill missing values for numeric columns with median values
df.fillna(df.select_dtypes(include=['number']).median(), inplace=True)

# Ensure no NaN values remain
if df.isnull().sum().sum() > 0:
    print("⚠️ Warning: Some NaN values still exist. Filling with 0.")
    df.fillna(0, inplace=True)

# Extract Song Names
song_names = df['Songsname']

# Extract Features (all numerical columns except "Songsname")
features = df.drop(columns=['Songsname'])

# Verify data types
print("\n🔍 Checking data types:\n", features.dtypes)
print("✅ Data Loaded & Prepared Successfully!")

# ------------------------ Train the KNN Model ------------------------
knn = NearestNeighbors(n_neighbors=6, metric='euclidean')  # 6 because one result will be the song itself
knn.fit(features)
print("✅ KNN Model Trained!")

# ------------------------ Song Recommendation Function ------------------------
def recommend_songs(song_name, df, knn, n_recommendations=5):
    """
    Recommend similar songs based on a given song name using the trained KNN model.
    """
    if song_name not in df['Songsname'].values:
        print(f"❌ Song '{song_name}' not found in dataset!")
        return []
    
    # Find the index of the given song
    song_index = df[df['Songsname'] == song_name].index[0]
    
    # Get the song's features
    song_features = features.iloc[[song_index]]  # Keep as DataFrame to preserve structure
    
    # Find the nearest neighbors
    distances, indices = knn.kneighbors(song_features, n_neighbors=n_recommendations + 1)
    
    # Get recommended songs (excluding the input song itself)
    recommended_songs = df.iloc[indices.flatten()[1:]]['Songsname'].tolist()
    
    return recommended_songs

# ------------------------ Test the Recommendation Function ------------------------
test_song = 'B56'  # Replace with any song name from your dataset
recommended_songs = recommend_songs(test_song, df, knn)

# Print Recommended Songs
print(f"\n🎵 Songs similar to '{test_song}':")
for song in recommended_songs:
    print(f"   - {song}")
```
